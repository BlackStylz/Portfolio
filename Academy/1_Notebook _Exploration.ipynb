# Filtrage des donn√©es

### Filtrage des pays

```python
p = pd.read_csv("./Data/Pays.csv") #Liste des pays reconnu par l'ONU
pays = p["Name"].tolist() 
```

```python
un_countries = countries.loc[countries["Short Name"].isin(pays)] # Filtrage grace √† la liste
un_countries.shape
```

Copie de data

```python
data_b = data.copy()
data_b.shape
```

Filtrage des pays sur la copie du fichier EdStatsData

```python
data_b = data_b.loc[data_b['Country Code'].isin(un_countries['Country Code'])]
data_b.shape
```

### Filtrage des indicateurs

Cr√©ation de la liste des indicateurs choisis historiques et pour la projection

```python
ind_hist = ["SP.POP.TOTL",
"SP.POP.1524.TO.UN",
"NY.GDP.PCAP.CD",
"NY.GNP.PCAP.CD",
"IT.NET.USER.P2", 
"UIS.GER.4", 
"UIS.CR.3",         
"SE.SEC.ENRR.UP",        
"UIS.SCHBSP.3.WCOMPUT",         
]

ind_prj = ["PRJ.ATT.15UP.4.MF",
"PRJ.MYS.15UP.MF",
"PRJ.ATT.15UP.3.MF"]
```

```python
data_fil = data_b.loc[data_b["Indicator Code"].isin(ind_hist) | data_b["Indicator Code"].isin(ind_prj)]
```

```python
data_fil.shape
```

Retirer les ann√©es avant 2005

```python
years=[]
for n in range(1970, 2005):
    years.append(str(n))
    
data_fil.drop(years,axis=1,inplace=True)

data_fil.shape
```

##### Dataframme donn√©es historiques

```python
df_hist = data_fil.loc[data_fil["Indicator Code"].isin(ind_hist)]
print(df_hist.shape)
msno.matrix(df_hist)
```

Supprimer les colonnes avec 10% ou plus de valeurs manquantes

```python
df_hist.dropna(thresh=(0.1*df_hist.shape[0]), axis=1, inplace =True)
msno.matrix(df_hist)
```

Cr√©er une table pivot

```python
country_bis = un_countries[['Country Code','Region']] # Extraire les r√©gions
```

```python
hist = df_hist.pivot_table(values='2016',index=['Country Code'] , columns=['Indicator Code'])
hist = pd.merge(hist ,country_bis, on="Country Code") #associer les r√©gions et les pays
hist.dropna(thresh =(0.5*hist.shape[0]), axis=1, inplace=True)
hist.dropna(inplace=True)
msno.matrix(hist)#.dropna(thresh=6, axis=0))
```

```python
hist.describe()
```

```python
#sns.boxplot(data=hist.groupby("Region").mean(), y="Region", x="NY.GDP.PCAP.CD")
plt.figure(figsize= (20,10))
plt.subplot(2,2,1)
sns.boxplot(
    data=hist, x="NY.GDP.PCAP.CD", y="Region",
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "coral"},
)
plt.xlabel("PIB par habitant ($)")

plt.subplot(2,2,2)
sns.boxplot(
    data=hist, x="NY.GNP.PCAP.CD", y="Region",
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "coral"},
)
plt.xlabel("PNB par habitant ($)")

plt.subplot(2,2,3)
sns.barplot(
    data=hist, x="SP.POP.TOTL", y="Region",
    linewidth=3, edgecolor=".5", facecolor=(.4, .6, .8, .5),
)
plt.xlabel("Population")

plt.subplot(2,2,4)
sns.boxplot(
    data=hist, x="IT.NET.USER.P2", y="Region",
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "coral"},
)
plt.xlabel("Utilisateur d'internet (%)")

plt.subplots_adjust(hspace=0.45,wspace=0.40)

plt.show()
```
