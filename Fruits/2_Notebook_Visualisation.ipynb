## Imports

```pyspark
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox, TextArea
from matplotlib.image import imread
```

#### Dimension dataFrame

```pyspark
pd.set_option('display.max_columns', 12)
```

### Chargement des données

```pyspark
PATH = 's3://p8datalubinstephane'
PATH_Result = PATH+'/Results'
df = pd.read_parquet(PATH_Result, engine='pyarrow')
```

```pyspark
#Séparation des composantes principales
for n in range(df.shape[0]):
    df.loc[n,'PC1'] = df.loc[n,'pcaFeatures']['values'][0]
    df.loc[n,'PC2'] = df.loc[n,'pcaFeatures']['values'][1]
```

### Visualisations

```pyspark
plt.figure(figsize=(35,35))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('PC1 (Composante Principale 1)',fontsize=30)
plt.ylabel('PC2 (Composante Principale 2)',fontsize=30)

for fruit in df.label.unique():
    dfred = df.loc[df['label']==fruit]
    plt.scatter(dfred['PC1'], dfred['PC2'], label=fruit, s=10)

plt.title("PCA - Projection sur PC1 et PC2",fontsize=40)

plt.legend()
#plt.savefig(PATH_Result+'/Visualisation_PCA.jpg')
%matplot plt
```
