#### Filtrage

Création d'un dataframme pour stocker les produits interéssants

Un tri est éffectué car certain produits n'ont pas d'image et d'autres n'ont pas les détails de la composition, on garde seulement les produits avec les 2

```python
df= pd.DataFrame(columns = ['foodId', 'label', 'category', 'foodContentsLabel', 'image'])
```

```python
for n in range(20):
    
    if ('foodContentsLabel' in response_dict['hints'][n]['food'].keys() ) & \
    ('image' in response_dict['hints'][n]['food'].keys() ):
        
        df.loc[n,'foodId'] = response_dict['hints'][n]['food']['foodId']
        df.loc[n,'label'] = response_dict['hints'][n]['food']['label']
        df.loc[n,'category'] = response_dict['hints'][n]['food']['category']
        df.loc[n,'foodContentsLabel'] = response_dict['hints'][n]['food']['foodContentsLabel']
        df.loc[n,'image'] = response_dict['hints'][n]['food']['image']
```
